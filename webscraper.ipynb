```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json

def get_google_search_paragraph(query):
    url = f"https://www.google.com/search?q={query}"
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        # The paragraph is usually contained within a div with class 'BNeawe'
        paragraph = soup.find('div', class_='BNeawe').text
        return paragraph
    else:
        print("Failed to retrieve search results.")
        return None
def load_descriptions(file_path):
    with open(file_path, 'r+') as file:
        descriptions = json.load(file)
    return descriptions
```

```python
data = pd.read_csv("compressed_dataset.csv")
desc_path = "desc.json"
descriptions = load_descriptions(desc_path)
```

```python

for idx, name in enumerate(data['name']):
    search_term = name + " Chemicals"
    if(search_term not in descriptions.keys()):
        top_paragraph = get_google_search_paragraph(search_term)
        if top_paragraph:
            descriptions[name] = top_paragraph
        if (idx + 1) % 10 == 0:
            with open(desc_path, 'w') as file:
                json.dump(descriptions, file)
            print(f"Descriptions saved for {idx + 1} terms.")

# Save remaining descriptions
if descriptions:
    with open(desc_path, 'w') as file:
        json.dump(descriptions, file)
```
